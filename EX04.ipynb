# **Relatório: Solução da Questão 1 - Regra de Simpson Composta**

## **Introdução**
Nesta questão, utilizamos um radar para medir a velocidade de um corredor durante os primeiros 5 segundos de uma corrida. Com os valores de velocidade medidos em intervalos regulares de tempo, a tarefa é estimar a distância percorrida pelo corredor ao longo desses 5 segundos utilizando a **Regra de Simpson Composta**.

A **Regra de Simpson** é uma técnica de integração numérica utilizada para estimar o valor de uma integral definida quando temos uma função discretizada em intervalos de tempo, ou seja, quando não conhecemos a expressão analítica da função, mas temos seus valores em pontos específicos.

## **Objetivo**
Estimar a distância que o corredor percorreu durante os 5 segundos usando a velocidade em função do tempo medida pela tabela abaixo:

\
\begin{array}{|c|c|}
\hline
t (\text{s}) & v (\text{m/s}) \\
\hline
0 & 4.67 \\
0.5 & 7.34 \\
1 & 8.86 \\
1.5 & 9.73 \\
2 & 10.22 \\
2.5 & 10.51 \\
3 & 10.67 \\
3.5 & 10.67 \\
4 & 10.81 \\
4.5 & 10.81 \\
5 & 10.81 \\
\hline
\end{array}


## **Metodologia**

### **Regra de Simpson Composta**
A **Regra de Simpson Composta** é uma técnica que divide o intervalo de integração em várias subpartes e aplica a Regra de Simpson a cada uma dessas subpartes.

A fórmula da Regra de Simpson Composta para integrar \( f(x) \) no intervalo \([a, b]\) com \(n\) subintervalos é dada por:

$$
\int_{a}^{b} f(x) dx \approx \frac{h}{3} \left[ f(x_0) + 4 \sum f(x_{\text{ímpar}}) + 2 \sum f(x_{\text{par}}) + f(x_n) \right]
$$

Onde:
- \(h\) é o tamanho do subintervalo, ou seja, $$h = \frac{b - a}{n}$$
- valor da função no primeiro ponto: $$f(x_0)$$
- valor da função no último ponto: $$f(x_n)$$
- referem-se aos valores da função nos pontos de índice ímpar os termos: $$f(x_{\text{ímpar}})$$
- referem-se aos valores da função nos pontos de índice par: $$f(x_{\text{par}})$$


### **Aplicação em Python**

Para resolver o problema, aplicamos o seguinte código Python para calcular a distância utilizando os valores fornecidos de tempo \( t \) e velocidade \( v(t) \).

```python
# Dados da tabela

v = np.array([4.67, 7.34, 8.86, 9.73, 10.22, 10.51, 10.67, 10.67, 10.81, 10.81, 10.81])  # Velocidade em m/s

    n = len(x) - 1  # Número de subintervalos
    if n % 2 == 1:
        raise ValueError("Número de subintervalos deve ser par.")
    h = (x[-1] - x[0]) / n
    soma = y[0] + y[-1]  # f(x0) + f(xn)
    
    # Soma dos termos ímpares
    soma_odd = sum(y[i] for i in range(1, n, 2))
    
    # Soma dos termos pares
    soma_even = sum(y[i] for i in range(2, n, 2))
    
    # Aplicando a fórmula de Simpson composta
    integral = (h / 3) * (soma + 4 * soma_odd + 2 * soma_even)
    return integral

# Calculando a distância
distancia = simpson_composta(t, v)

print(f"A distância estimada que o corredor cobriu é: {distancia:.4f} metros")

# **Relatório: Solução da Questão 2 - Regra dos Trapézios Generalizada**

## **Introdução**
Nesta questão, o problema envolve o cálculo da área sob uma curva que descreve a variação da pressão ao longo de uma parede de barragem, variando com a profundidade $h$ Sabemos que a função que descreve essa relação é:

$$
P(h) = 5h^2 + 3h + 2
$$

Deseja-se estimar a área total sob essa curva utilizando a **Regra dos Trapézios Generalizada**. A área estimada representará a pressão total sobre a parede da barragem. O intervalo de profundidade vai de 0 a 10 metros, e o intervalo deve ser dividido em 5 segmentos iguais.

## **Objetivo**
Estimar a pressão total sobre a parede da barragem entre 0 e 10 metros de profundidade, dividindo o intervalo em 5 segmentos iguais, e aplicando a **Regra dos Trapézios Generalizada** para essa estimativa.

## **Metodologia**

### **Regra dos Trapézios Generalizada**
A **Regra dos Trapézios** é um método de integração numérica que aproxima a área sob uma curva dividindo o intervalo em segmentos e calculando a área de trapézios formados pelos pontos no gráfico da função. A fórmula da Regra dos Trapézios Generalizada para integrar uma função f(x) no intervalo [a, b] com n subintervalos é:

$$
\int_{a}^{b} f(x) dx \approx \frac{h}{2} \left[ f(x_0) + 2 \sum f(x_i) + f(x_n) \right]
$$

Onde:
- h é o tamanho do subintervalo, dado por $$h = \frac{b - a}{n}$$
- $f(x_0)$ e $f(x_n)$ são os valores da função no início e no final do intervalo;
- soma dos valores da função nos pontos intermediários: $$\sum f(x_i)$$

### **Aplicação em Python**

Para resolver essa questão, podemos implementar o método da **Regra dos Trapézios Generalizada** em Python e aplicá-lo à função dada \(P(h)\). O código a seguir faz isso:

```python
# Função que define a pressão em função da profundidade
def P(h):
    return 5 * h**2 + 3 * h + 2

# Definindo os parâmetros do problema
a = 0  # Profundidade inicial
b = 10  # Profundidade final
n = 5  # Número de subintervalos (dividir em 5 segmentos iguais)
h = (b - a) / n  # Tamanho de cada subintervalo

# Calculando a Regra dos Trapézios
x = np.linspace(a, b, n+1)  # Pontos que dividem o intervalo
y = P(x)  # Valores da função em cada ponto

# Aplicação da Regra dos Trapézios Generalizada
def regra_trapezios(x, y, h):
    soma = y[0] + y[-1]  # f(x0) + f(xn)
    soma += 2 * sum(y[1:-1])  # 2 * somatório dos termos intermediários
    integral = (h / 2) * soma
    return integral

# Calculando a pressão total
pressao_total = regra_trapezios(x, y, h)

print(f"A pressão total estimada sobre a parede da barragem é: {pressao_total:.4f} unidades")
```

## **Descrição do Código**

1. **Função P(h):** Implementa a função fornecida $P(h) = 5h^2 + 3h + 2$, que define a pressão em função da profundidade.

2. **Parâmetros:** Definimos o intervalo de profundidade entre \( a = 0 \) e \( b = 10 \), com \( n = 5 \) subintervalos. A largura de cada subintervalo \(h\) é calculada como:

$$
h = \frac{b - a}{n}
$$

3. **Aplicação da Regra dos Trapézios**: A função `regra_trapezios` implementa a fórmula da **Regra dos Trapézios Generalizada**, somando os valores dos pontos extremos e intermediários conforme a fórmula.

4. **Cálculo da pressão total**: Após aplicar a fórmula, o valor obtido representa a pressão total estimada sobre a parede da barragem.


````python
# Função que define a pressão em função da profundidade
def P(h):
    return 5 * h**2 + 3 * h + 2

# Definindo os parâmetros do problema
a = 0  # Profundidade inicial
b = 10  # Profundidade final
n = 5  # Número de subintervalos (dividir em 5 segmentos iguais)
h = (b - a) / n  # Tamanho de cada subintervalo

# Calculando a Regra dos Trapézios
x = np.linspace(a, b, n+1)  # Pontos que dividem o intervalo
y = P(x)  # Valores da função em cada ponto

# Aplicação da Regra dos Trapézios Generalizada
def regra_trapezios(x, y, h):
    soma = y[0] + y[-1]  # f(x0) + f(xn)
    soma += 2 * sum(y[1:-1])  # 2 * somatório dos termos intermediários
    integral = (h / 2) * soma
    return integral

# Calculando a pressão total
pressao_total = regra_trapezios(x, y, h)

print(f"A pressão total estimada sobre a parede da barragem é: {pressao_total:.4f} unidades")
````

# **Relatório: Solução da Questão 3b - Método de Gauss-Seidel**

## **Introdução**
Nesta questão, foi proposto um sistema de equações lineares que representa o balanceamento de carga em diferentes nós de uma rede de energia elétrica. Nosso objetivo é resolver o sistema utilizando o **Método de Gauss-Seidel**, com uma tolerância (&epsilon; = 0,001). O sistema de equações é dado por:

\
\begin{cases}
3x_1 - x_2 + x_3 = 1 \\
3x_1 + 3x_2 + 7x_3 = 4 \\
3x_1 + 6x_2 + 2x_3 = 0 \\
\end{cases}


## **Objetivo**
Resolver o sistema de equações lineares utilizando o método iterativo de **Gauss-Seidel** até que a diferença entre iterações consecutivas seja menor que \( &epsilon; = 0,001 \).

## **Metodologia**

### **Método de Gauss-Seidel**
O **Método de Gauss-Seidel** é um método iterativo para resolver sistemas de equações lineares. Nele, as equações são reescritas de forma que cada variável $x_i$ seja isolada, e em cada iteração, os valores mais recentes das variáveis são utilizados para calcular os novos valores.

O sistema pode ser reescrito da seguinte forma:

$$
[
x_1 = \frac{1 + x_2 - x_3}{3}
]
$$

$$
[
x_2 = \frac{4 - 3x_1 - 7x_3}{3}
]
$$

$$
[
x_3 = \frac{-3x_1 - 6x_2}{2}
]
$$


### **Aplicação em Python**

O código a seguir implementa o método de Gauss-Seidel para resolver o sistema proposto:

```python

# Função para rearranjar a matriz para garantir a dominância diagonal
def rearrange_matrix(A, b):
    """
    Rearranja a matriz A e o vetor b para garantir que os maiores valores absolutos
    estejam na diagonal principal (para melhor desempenho no método de Gauss-Seidel).
    """
    n = len(b)
    for i in range(n):
        max_index = np.argmax(np.abs(A[i:, i])) + i
        if i != max_index:
            A[[i, max_index]] = A[[max_index, i]]
            b[[i, max_index]] = b[[max_index, i]]
    return A, b

# Definir a função do método de Gauss-Seidel
def gauss_seidel(A, b, tol=1e-3, max_iterations=1000):
    """
    Método de Gauss-Seidel para resolver sistemas de equações lineares.
    
    Parâmetros:
    A: Matriz de coeficientes
    b: Vetor de constantes
    tol: Tolerância para o critério de parada
    max_iterations: Número máximo de iterações
    
    Retorna:
    x: Vetor solução
    """
    n = len(b)
    x = np.zeros_like(b, dtype=np.float64)  # Vetor inicial de zeros
    for iteration in range(max_iterations):
        x_new = np.copy(x)
        for i in range(n):
            # Calcular a soma das partes anteriores e posteriores
            s1 = np.dot(A[i, :i], x_new[:i])
            s2 = np.dot(A[i, i + 1:], x[i + 1:])
            x_new[i] = (b[i] - s1 - s2) / A[i, i]
        # Critério de parada
        if np.allclose(x, x_new, atol=tol, rtol=0.):
            break
        x = x_new
    return x

# Definir a matriz dos coeficientes e o vetor de constantes
A = np.array([[3, -1, 1], [3, 3, 7], [3, 6, 2]], dtype=np.float64)
b = np.array([1, 4, 0], dtype=np.float64)

# Rearranjar a matriz para garantir a dominância diagonal
A, b = rearrange_matrix(A, b)

# Chamar a função com tolerância de 0,001
solucao = gauss_seidel(A, b, tol=1e-3)

print("Solução encontrada pelo Método de Gauss-Seidel:")

# **Relatório: Solução da Questão 4 - Método de Resolução de Sistemas Lineares**

## **Introdução**
Nesta questão, temos um sistema de equações lineares que representa a produção de quatro tipos de computadores. Cada computador exige uma certa quantidade de mão-de-obra, metais, plásticos e componentes eletrônicos. Nosso objetivo é determinar quantos computadores de cada tipo podem ser produzidos diariamente, dado o limite de recursos disponível.

O sistema de equações foi construído com base nos seguintes coeficientes para cada recurso:

$$
\begin{array}{c|c|c|c|c}
\text{Tipo de Computador} & \text{Mão de Obra (h/comp.)} & \text{Metais (Kg/comp.)} & \text{Plásticos (Kg/comp.)} & \text{Componentes (unid./comp.)} \\
\hline
1 & 3 & 20 & 10 & 10 \\
2 & 4 & 25 & 15 & 8 \\
3 & 7 & 40 & 20 & 10 \\
4 & 20 & 50 & 22 & 15 \\
\end{array}
$$

## **Objetivo**
Resolver o sistema de equações lineares que representa os limites diários de produção de computadores para os quatro tipos, dados os seguintes recursos disponíveis:

- **Mão de obra**: 504 horas por dia.
- **Metais**: 1970 Kg por dia.
- **Plásticos**: 970 Kg por dia.
- **Componentes**: 601 unidades por dia.

## **Metodologia**

### **Sistema de Equações**
O sistema de equações lineares que representa o consumo de recursos para cada tipo de computador é:

\
\
\begin{cases}
3x_1 + 4x_2 + 7x_3 + 20x_4 = 504 & \text{(Mão de obra)} \\
20x_1 + 25x_2 + 40x_3 + 50x_4 = 1970 & \text{(Metais)} \\
10x_1 + 15x_2 + 20x_3 + 22x_4 = 970 & \text{(Plásticos)} \\
10x_1 + 8x_2 + 10x_3 + 15x_4 = 601 & \text{(Componentes)} \\
\end{cases}

Resolvendo de maneira analítica, os respecitos valores de $x_1$, $x_2$, $x_3$, $x_4$ são: 10, 12, 18, 15. Vamos utilizar o método de Gauss Seidel para resolver de modo numérico.

### **Aplicação em Python**

O código a seguir implementa o método de Gauss Seidel para resolver o sistema proposto:

```python
# Função de Gauss-Seidel
def gauss_seidel(A, b, x0, tol=1e-5, max_iteracoes=100):
    n = len(b)
    x = x0.copy()
    
    for iteracao in range(max_iteracoes):
        x_novo = x.copy()
        
        for i in range(n):
            soma_Ax = sum(A[i][j] * x_novo[j] for j in range(n) if j != i)
            x_novo[i] = (b[i] - soma_Ax) / A[i][i]
        
        # Verificar convergência
        if np.linalg.norm(x_novo - x, ord=np.inf) < tol:
            return x_novo, iteracao + 1
        
        x = x_novo.copy()
    
    return x, max_iteracoes

# Verificar se a matriz A é estritamente diagonal dominante
def matriz_diagonal_dominante(A):
    n = A.shape[0]
    for i in range(n):
        soma_linha = sum(abs(A[i][j]) for j in range(n) if j != i)
        if abs(A[i][i]) <= soma_linha:
            return False
    return True

# Verificar critério de Sassenfeld
def criterio_sassenfeld(A):
    n = A.shape[0]
    beta = np.zeros(n)
    
    for i in range(n):
        soma_antes = sum(abs(A[i][j]) * beta[j] for j in range(i))
        soma_depois = sum(abs(A[i][j]) for j in range(i + 1, n))
        beta[i] = (soma_antes + soma_depois) / abs(A[i][i])
        
    return beta, max(beta) < 1

# Coeficientes das equações
A = np.array([
    [3, 4, 7, 20],  # Equação 1 - Mão-de-obra
    [20, 25, 40, 50],  # Equação 2 - Metais
    [10, 15, 20, 22],  # Equação 3 - Plásticos
    [10, 8, 10, 15]   # Equação 4 - Componentes
])

# Recursos disponíveis (lado direito das equações)
b = np.array([504, 1970, 970, 601])

# Chute inicial (valores iniciais de x1, x2, x3, x4)
x0 = np.zeros(4)

# Verificar se A é diagonal dominante
diagonal_dominante = matriz_diagonal_dominante(A)

# Aplicar critério de Sassenfeld
valores_beta, sassenfeld_converge = criterio_sassenfeld(A)

Portanto, não foi possível aplicar o método de Gauss-Seidel para este sistema, devido à falta de convergência garantida pelos critérios matemáticos.

````python
# Função de Gauss-Seidel
def gauss_seidel(A, b, x0, tol=1e-5, max_iteracoes=100):
    n = len(b)
    x = x0.copy()

    for iteracao in range(max_iteracoes):
        x_novo = x.copy()

        for i in range(n):
            soma_Ax = sum(A[i][j] * x_novo[j] for j in range(n) if j != i)
            x_novo[i] = (b[i] - soma_Ax) / A[i][i]

        # Verificar convergência
        if np.linalg.norm(x_novo - x, ord=np.inf) < tol:
            return x_novo, iteracao + 1

        x = x_novo.copy()

    return x, max_iteracoes

# Verificar se a matriz A é estritamente diagonal dominante
def matriz_diagonal_dominante(A):
    n = A.shape[0]
    for i in range(n):
        soma_linha = sum(abs(A[i][j]) for j in range(n) if j != i)
        if abs(A[i][i]) <= soma_linha:
            return False
    return True

# Verificar critério de Sassenfeld
def criterio_sassenfeld(A):
    n = A.shape[0]
    beta = np.zeros(n)

    for i in range(n):
        soma_antes = sum(abs(A[i][j]) * beta[j] for j in range(i))
        soma_depois = sum(abs(A[i][j]) for j in range(i + 1, n))
        beta[i] = (soma_antes + soma_depois) / abs(A[i][i])

    return beta, max(beta) < 1

# Coeficientes das equações
A = np.array([
    [3, 4, 7, 20],  # Equação 1 - Mão-de-obra
    [20, 25, 40, 50],  # Equação 2 - Metais
    [10, 15, 20, 22],  # Equação 3 - Plásticos
    [10, 8, 10, 15]   # Equação 4 - Componentes
])

# Recursos disponíveis (lado direito das equações)
b = np.array([504, 1970, 970, 601])

# Chute inicial (valores iniciais de x1, x2, x3, x4)
x0 = np.zeros(4)

# Verificar se A é diagonal dominante
diagonal_dominante = matriz_diagonal_dominante(A)

# Aplicar critério de Sassenfeld
valores_beta, sassenfeld_converge = criterio_sassenfeld(A)

# Tentar resolver usando Gauss-Seidel
if diagonal_dominante or sassenfeld_converge:
    resultado_x, iteracoes = gauss_seidel(A, b, x0)
    motivo_falha = None
else:
    resultado_x = None
    iteracoes = 0
    motivo_falha = "Critérios de convergência não foram satisfeitos: nem Dominância Diagonal, nem Critério de Sassenfeld."

# Exibir resultado de forma mais amigável
def exibir_resultado():
    print("Resultado do Método de Gauss-Seidel\n")
    print(f"Dominância Diagonal: {'Sim' if diagonal_dominante else 'Não'}")
    print(f"Critério de Sassenfeld atendido: {'Sim' if sassenfeld_converge else 'Não'}")
    print("\nValores de β (Critério de Sassenfeld):")
    for i, beta in enumerate(valores_beta):
        print(f"β{i+1}: {beta:.4f}")

    if motivo_falha:
        print(f"\nFalha: {motivo_falha}")
    else:
        print(f"\nMétodo convergiu após {iteracoes} iterações.")
        print("Valores obtidos:")
        for i, valor in enumerate(resultado_x):
            print(f"x{i+1}: {valor:.4f}")

# Exibir o resultado no terminal
exibir_resultado()
````
