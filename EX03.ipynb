## Resolução da questão

A equação diferencial dada é:

$$
2x + yy' = y^2
$$

Com a condição inicial \( y(0) = 1 \). A solução exata é fornecida como:

$$
y(x) = \sqrt{2x + 1}
$$

O objetivo é calcular a aproximação de \( y(1) \) usando os dois métodos numéricos com passo \( h = 0,2 \), e comparar as soluções aproximadas com a solução exata para determinar o erro absoluto.

## Definir a Função da EDO

A equação diferencial $$( 2x + yy' = y^2 )$$ foi rearranjada para resolver \( y' \):

$$
y' = y - \frac{2x}{y}
$$

Esta equação é representada no código como:

```python
def f(x, y):
    return y - (2 * x / y)
```

## Método de Euler

O método de Euler é uma aproximação numérica simples para resolver equações diferenciais. A fórmula geral do método de Euler para um passo \( h \) é dada por:

$$
y_{n+1} = y_n + h \cdot f(x_n, y_n)
$$

No código, o método de Euler foi implementado como:

```python
def euler(f, xn, yn, h):
    return yn + h * f(xn, yn)
```

## Método de Runge-Kutta de 3ª Ordem

O método de Runge-Kutta de 3ª ordem oferece uma maior precisão em relação ao método de Euler. Ele utiliza três avaliações da função \( f(x, y) \) em diferentes pontos para calcular uma média ponderada dos incrementos. A fórmula geral é:

$$
k_1 = f(x_n, y_n)
$$

$$
k_2 = f\left(x_n + \frac{h}{2}, y_n + \frac{h}{2} \cdot k_1\right)
$$

$$
k_3 = f\left(x_n + h, y_n - h \cdot k_1 + 2h \cdot k_2\right)
$$

E a aproximação de $$( y_{n+1} )$$ é dada por:

$$
y_{n+1} = y_n + \frac{h}{6} \left(k_1 + 4k_2 + k_3\right)
$$

No código, o método de Runge-Kutta foi implementado da seguinte forma:

```python
def runge_kutta3(f, xn, yn, h):
    k1 = f(xn, yn)
    k2 = f(xn + h / 2, yn + k1 * h / 2)
    k3 = f(xn + h / 2, yn + h / 2 * (k1 + k2) / 2)
    return yn + h * (k1 + 4 * k2 + k3) / 6
```

## Comparação com a Solução Exata

Sabemos que a solução exata da equação diferencial é:

$$
y(x) = \sqrt{2x + 1}
$$

O código calcula a solução exata para os mesmos valores de \( x \) usados nos métodos numéricos:

```python
def solve_exato(f, xn, yn, h):
    return np.sqrt(2 * (xn + h) + 1)
```

Cálculo do Erro Absoluto
O erro absoluto é calculado comparando os valores aproximados de y(1) obtidos pelos métodos numéricos com a solução exata.

Código para calcular o erro absoluto para os métodos de Euler e Runge-Kutta é:

```python
print(f"Erro absoluto de Runge-Kutta = {np.abs(yrk[i] - yexato[i])}")
print(f"Erro absoluto de Euler = {np.abs(yeuler[i] - yexato[i])}")
```

````python
#--------------------------PARAMETROS-------------------------------------------------
xspan = [0,1]
y0 = 1
h = 0.2

#---------------------------------Manipulacao da PVI----------------------------------
#resolver a PVI com Euler ou Ruge Kutta dependendo da funcao que for passada para ele
def solve_pvi(f,xspan,y0,h,solver):
    xs = np.arange(xspan[0] , xspan[1] +  h, h)# Definir os valores de x
    ys = np.zeros(len(xs))#cria os y começando com 0
    ys[0] = y0

    for i in range(len(xs) - 1):
        ys[i+1] = solver(f, xs[i], ys[i], h)

    return xs, ys

#definir a  PVI para y(x):
def f(x,y):
    return y-(2*x/y)


#-------------------------------------EULER-----------------------------------------
#metodo euler:
def euler(f,xn, yn, h ):
    return yn + h * f(xn,yn)


#-----------------------------METODO DE RUNGE-KUTTA 3º ORDEM------------------------


def runge_kutta3(f,xn, yn,h):
        k1 =f(xn,yn)
        k2 =f(xn + h/2 , yn + k1 * h/2)
        k3 = f(xn + h/2, yn + (h/2) * (k1+k2)/2)
        return (yn +  h * (k1 + 4*k2 + k3)/6)


#------------------------------------VALOR EXATO------------------------------------
# Resolver a PVI diretamente com a solução exata
def solve_exato(f,xn,yn,h):
    return np.sqrt(2 * (xn + h) + 1)




 #--------------------------------Printar Resultados----------------------------------
from tabulate import tabulate

# Resolver PVI usando o método de Euler, Runge-Kutta, e a solução exata
xeuler, yeuler = solve_pvi(f, xspan, y0, h, euler)
xexato, yexato = solve_pvi(f, xspan, y0, h, solve_exato)
xrk, yrk = solve_pvi(f, xspan, y0, h, runge_kutta3)

# Criar dados para a tabela
data = []
for i in range(len(xrk)):
    erro_rk = np.abs(yrk[i] - yexato[i])
    erro_euler = np.abs(yeuler[i] - yexato[i])
    data.append([xrk[i], yeuler[i], yrk[i], yexato[i], erro_euler, erro_rk])

# Definir os nomes das colunas
col_names = ["x", "Euler", "Runge-Kutta 3", "Exato", "Erro Euler", "Erro Runge-Kutta"]

# Exibir tabela
print(tabulate(data, headers=col_names, floatfmt=".5f"))
````

##Definição da Equação Diferencial

$$

Onde a força da gravidade $$( F_g = m g )$$ e a resistência do ar é modelada como $$( F_r = - k v |v| )$$

```python
def dv_dt(t, v):
```

##Método de Runge-Kutta de 4ª Ordem

O método de Runge-Kutta de quarta ordem é utilizado para resolver numericamente a equação diferencial. Ele calcula quatro coeficientes $$( k_1, k_2, k_3 ),(k_4)$$ para cada passo de tempo, e a fórmula geral é dada por:

$$
k_1 = h f(t_n, v_n)
$$

$$
k_2 = h f(t_n + \frac{h}{2}, v_n + \frac{k_1}{2})
$$

$$
k_3 = h f(t_n + \frac{h}{2}, v_n + \frac{k_2}{2})
$$

$$
k_4 = h f(t_n + h, v_n + k_3)
$$

A próxima aproximação da velocidade $$( v_{n+1} )$$ é então calculada por:

$$
v_{n+1} = v_n + \frac{k_1 + 2k_2 + 2k_3 + k_4}{6}
$$

No código, o método de Runge-Kutta foi implementado como:

```python
def runge_kutta4(f, t0, v0, h):
    k1 = h * f(t0, v0)
```

##Questão 3: Crescimento Populacional

##Resolução:

Seja \( P(t) \) o número de indivíduos de uma certa população medido em anos. Sabemos que:

*   A taxa de nascimentos \( b \) é constante.
*   A taxa de mortalidade \( d \) é proporcional ao tamanho da população, ou seja, \( d = kP(t) \).

Isso nos leva à equação logística para o crescimento populacional:

$$
\frac{dP(t)}{dt} = bP(t) - kP(t)^2
$$

-

$$( P(0) = 50976 )$$
$$( b = 2.9 \times 10^{-2})$$
$$( k = 1.4 \times 10^{-7})$$
$$( t = 5 ) anos.$$

O objetivo é encontrar a população estimada após 5 anos utilizando o método de Runge-Kutta de quarta ordem.

##Código Implementado

O seguinte código em Python foi utilizado para resolver o problema:

#---PARAMETROS---
```python
p0 = 50976  # Populacao inicial
b = 2.9 * 10**-2  # Taxa de nascimentos
k = 1.4 * 10**-7  # Constante de mortalidade proporcional
t_final = 5  # Tempo final (em anos)
h = 0.05  # Passo para o método de Runge-Kutta (5/0.05 = 100 repeticoes)
```

#---Definir o crescimento da população---

```python
# Definindo a funcao da derivada dp/dt
def dp_dt(_t, p):

```

#---Método de Runge-Kutta de 4ª ordem---

```python
# Implementação do método de Runge-Kutta de 4ª ordem
def runge_kutta4(f, tn, pn, h):
    k1 = h * f(tn, pn)
    k2 = h * f(tn + h / 2, pn + k1 / 2)
    k3 = h * f(tn + h / 2, pn + k2 / 2)
    k4 = h * f(tn + h, pn + k3)
    return pn + (k1 + 2 * k2 + 2 * k3 + k4) / 6
```


#---Resolucao da PVI---
```python
# Funcao para resolver a PVI (problema de valor inicial)
def solve_pvi(f, t0, t_final, p0, h, solver):
```

##Função de crescimento da população:

A derivada $$( \frac{dP}{dt} )$$ é definida como:     

$$(\frac{dP(t)}{dt} = bP(t) - kP(t)^2)$$

Essa equação é implementada na função `dp_dt`.

##Método de Runge-Kutta de quarta ordem:

O método de Runge-Kutta de quarta ordem é implementado para resolver a equação diferencial com um passo \( h \), calculando o valor de \( P(t) \) em cada ponto \( t \).
    
    
##Solução numérica:
A função `solve_pvi` é usada para resolver o problema de valor inicial (PVI), retornando os valores de \( P(t) \) em função do tempo \( t \).

##Resultados:

O código imprime a população para cada tempo \( t \) e mostra a população estimada após 5 anos.  
    
Utilizando o método de Runge-Kutta de quarta ordem, a população estimada depois de 5 anos é exibida ao final do código. Esse método é eficaz para resolver equações diferenciais como essa, que modela o crescimento de uma população com base em uma taxa de nascimento e uma mortalidade proporcional à população.

````python
#--------------------------PARAMETROS-------------------------------------------------
p0 = 50976  # População inicial
b = 2.9 * 10**-2  # Taxa de nascimentos
k = 1.4 * 10**-7  # Taxa de mortalidade proporcional
t_final = 5  # Tempo final
h = 0.05  # Passo de integração (h)

#---------------------------------Manipulacao da PVI----------------------------------
# Resolver a PVI
def solve_pvi(f, t0, t_final, p0, h, solver):
    n = int((t_final - t0) / h)  # Número de iterações
    tabela_t = np.arange(t0, t_final + h, h)  # Definir os valores de t
    tabela_p = np.zeros(len(tabela_t))  # Criar a tabela de população começando com 0
    tabela_p[0] = p0

    for i in range(n):
        tabela_p[i + 1] = solver(f, tabela_t[i], tabela_p[i], h)

    return tabela_t, tabela_p

# Definir o crescimento da população
def dp_dt(_t, p):
    return b * p - k * p**2

#--------------------------------Runge Kutta 4 ------------------------------------
# Implementação do método de Runge-Kutta de 4ª ordem
def runge_kutta4(f, tn, pn, h):
    k1 = h * f(tn, pn)
    k2 = h * f(tn + h / 2, pn + k1 / 2)
    k3 = h * f(tn + h / 2, pn + k2 / 2)
    k4 = h * f(tn + h, pn + k3)
    return pn + (k1 + 2 * k2 + 2 * k3 + k4) / 6

#--------------------------------------Resolucao-----------------------------------
# Resolver a PVI com o método de Runge-Kutta 4
xrk4, yrk4 = solve_pvi(dp_dt, 0, t_final, p0, h, runge_kutta4)

# Exibir os valores calculados
for i in range(len(xrk4)):
    print(f"x = {xrk4[i]:.3f} | y = {yrk4[i]:.4f}")

print(f"\nPopulação esperada depois de 5 anos é: {yrk4[-1]:.2f}\n")

#---------------------------------Plotar o Gráfico---------------------------------
plt.plot(xrk4, yrk4, label="População ao longo do tempo", color='blue')
plt.xlabel('Tempo (anos)')
plt.ylabel('População')
plt.title('Crescimento Populacional - Método de Runge-Kutta 4ª ordem')
plt.legend()
plt.grid(True)
plt.show()
````
